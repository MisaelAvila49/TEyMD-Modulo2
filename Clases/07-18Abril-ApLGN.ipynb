## El problema de la aguja de Bufón


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parametros del problema
# Distancia entre las lineas paralelas de 1
# Longitud de la aguja de 1
np.random.seed(42) # Para reproducibilidad
L = 1 # Longitud de la aguja
d = 1 # Distancia entre las lineas paralelas
num_lanzamientos = 500000

num_pasos = 1000

# Creamos listas vacías para guardar los resultados
pi_estimados = []
lanzamientos_realizados = []

# Contamos los cruces de la línea para saber cuantas agujas cruzan una línea
cruces = 0

# Simulación de los lanzamientos de la aguja y la estimación de pi
for i in range(1,num_lanzamientos+1):
    # Posición del centro de la aguja
    centro = np.random.uniform(0, d/2)

    # Ángulo aleatorio entre 0 y pi
    angulo = np.random.uniform(0, np.pi)

    # Distancia desde el centro de la aguja hasta la línea más cercana
    distancia = (L / 2) * np.sin(angulo)

    # Verificamos si la aguja cruza una línea
    if centro <= distancia:
        cruces += 1 #Incrementar el contador de cruces

    # Estimamos pi usando la fórmula de Buffon
    if i % num_pasos == 0 and cruces > 0: # Solo estimamos pi cada 'step' lanzamientos y si hay al menos un cruce
        pi_estimado = (2 * L * i) / (d * cruces)
        pi_estimados.append(pi_estimado)
        lanzamientos_realizados.append(i)
        
print(f'Estimación de pi: {pi_estimados}')      

# Graficamos la estimación de pi a lo largo de los lanzamientos
plt.figure(figsize=(10, 6))
plt.plot(lanzamientos_realizados, pi_estimados, label='Estimación de pi', color='blue')
plt.axhline(y=np.pi, color='red', linestyle='--', label='Valor real de pi')
plt.xlabel('Número de lanzamientos')
plt.ylabel('Estimación de pi')
plt.legend()
plt.show()
```
